```python
import numpy as np

n = 3
extra = np.array([[1/2, 1/2, 0], [1/2, 0, 1], [0, 1/2, 0]])
print("Matrix:\n",extra)
limit = 0.000000001
rj = np.array([1/n,1/n,1/n])
for _ in range(100):
  new_rj = np.dot(extra, rj)
  print(new_rj)
  if np.linalg.norm(new_rj - rj) < limit:
    break
  rj = new_rj
print("\nFinal Matrix:\n",rj)
```

```python
# Google Matrix A = ß * M + (1 -ß) [1/N]

import numpy as np

n = 3
beta = 0.8
M = np.array([[1/2, 1/2, 0], [1/2, 0, 0], [0, 1/2, 1]])
A = beta*M + (1-beta)*np.full((3, 3), 1/n)
print("\nMatrix M:\n",M)
print("\nMatrix A:\n",A)
limit = 0.00000000000000000001
rj = np.array([1/n,1/n,1/n])
while True:
  new_rj = np.dot(A, rj)
  print(new_rj)
  if np.linalg.norm(new_rj - rj) < limit:
    break
  rj = new_rj
print("\nFinal Matrix:\n",rj)
```
